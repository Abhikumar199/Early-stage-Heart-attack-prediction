```python
import pandas as pd
import numpy as np

# Assume your dataframe is df and column name is 'WEIGHT2'

# Make a copy
df['WEIGHT_KG'] = df['WEIGHT2']

# Step 1: Convert pounds (50â€“776) to kg
mask_pounds = (df['WEIGHT2'] >= 50) & (df['WEIGHT2'] <= 776)
df.loc[mask_pounds, 'WEIGHT_KG'] = df.loc[mask_pounds, 'WEIGHT2'] * 0.453592

# Step 2: Convert metric-coded weights (9023â€“9352) to proper kg
# These are stored as 9xxx, so remove the leading 9000 part
mask_metric = (df['WEIGHT2'] >= 9023) & (df['WEIGHT2'] <= 9352)
df.loc[mask_metric, 'WEIGHT_KG'] = df.loc[mask_metric, 'WEIGHT2'] - 9000

# Step 3: Replace invalid/missing codes with NaN
df.loc[df['WEIGHT2'].isin([7777, 9999]), 'WEIGHT_KG'] = np.nan

# Optional: drop or handle missing
df = df.dropna(subset=['WEIGHT_KG'])
```
